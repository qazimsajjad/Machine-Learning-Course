## Implementation of Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(trainImages):
    plt.figure(figsize=(10,10))
    _, axis = plt.subplots(4,4)
    for i, ax in enumerate(axis.ravel()):
        ax.imshow(trainImages[i], 'gray')
        ax.axis('off')
    plt.show()
        
def visualize(x, y, functionName):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=functionName, color='blue')
    plt.title(f'{functionName.title()} Activation Function')
    plt.xlabel('x')
    plt.ylabel(f'{functionName.title()}(x)')
    plt.axhline(0.5, color='red', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()
```

# Activation Functions:
image.png

```python
def ourSigmoid(x):
    return 1/(1+np.exp(-x))
x = np.linspace(-10, 10, 100)
y = ourSigmoid(x)
visualize(x, y, 'sigmoid')
```

# ReLU Activation Function:

image.png

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)
visualize(x, y, 'ReLU')
```

# LeakyReLU Activation Function:
 ##  considering alpha times of negative.
image.png

```python
def leakyReLU(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

x = np.linspace(-10, 10, 100)
y = leakyReLU(x)
visualize(x, y, 'leakyReLU')
```

# Tangent Hyperbolic activation Function.

image.png

```python
def ourTanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
x = np.linspace(-10, 10, 100)
y = ourTanh(x)
visualize(x, y, 'tanh')
```

# Scaled Exponential Linear Unit (SELU)

image.png

```python
def ourSeLU(x, alpha=1.67326324 , scale=1.05070098):
    return np.where(x<0, alpha*scale*(np.exp(x)-1), scale*x)
x = np.linspace(-10, 10, 100)
y2 = ourSeLU(x)
visualize(x, y2, 'ourselu')
```

# Softmax Function


```python
def ourSoftmax(x):
    return np.exp(x) / np.sum(np.exp(x))
x = np.linspace(-10, 10, 100) 
y = ourSoftmax(x)
visualize(x,y,'softmax')
```

# Exponential Linear Unit (ELU)


```python
def ourelu(x, alpha=1.0):
    return np.where(x <= 0, alpha * (np.exp(x) - 1), x)
x = np.linspace(-4, 4, 100)
y = ourelu(x)
visualize(x, y, 'elu')
```

# Softplus Function


```python
def ourSoftplus(x):
    return np.log(1+np.exp(x))
x = np.linspace(-6, 6, 100)
y = ourSoftplus(x)
visualize(x, y, 'softplus')
```

# Swish Function


```python
def ourSwish(x):
    return x*ourSigmoid(x)
x = np.linspace(-4, 4, 100)
y = ourSwish(x)
visualize(x, y, 'swish')
```
